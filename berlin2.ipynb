```python
# Importing the required module for regression tasks from the pycaret library
from pycaret.regression import *
```

```python
df.columns
```

```python
# Creating a copy of the original DataFrame for prediction purposes
df_pred = df.copy()

# Selecting a subset of columns for the prediction DataFrame
df_pred_new = df_pred[["neighbourhood_group", "accommodates", "property_type", "room_type", "price", "minimum_nights", "maximum_nights"]]
```

```python
df_pred_new.head()
```

```python
# Setting up the PyCaret environment for regression with the selected DataFrame and target variable
setup(df_pred_new, target="price")
```

```python
# Comparing various regression models and selecting the best-performing one
best = compare_models()
```

```python
# Creating a LightGBM regression model
lightgbm = create_model("lightgbm")
```

```python
# Tuning hyperparameters of the LightGBM model
tuned_lightgbm = tune_model(lightgbm)
```

```python
# Evaluating the performance of the tuned LightGBM model
evaluate_model(tuned_lightgbm)
```

```python
final_lightgbm_best = finalize_model(tuned_lightgbm)
```

```python
save_model(final_lightgbm_best, "ml_airbnb")
```

```python
# Creating a xgboost regression model
xgboost = create_model("xgboost")
```

```python
# Tuning hyperparameters of the xgboost model
tuned_xgboost = tune_model(xgboost)
```

```python
# Evaluating the performance of the tuned xgboost model
evaluate_model(tuned_xgboost)
```

```python
final_xgboost_best = finalize_model(tuned_xgboost)
```

```python
save_model(final_xgboost_best, "ml_airbnb_xgb")
```
